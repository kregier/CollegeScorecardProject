Based on U.S. Department of Education, National Center for Education Statistics. (2019). Digest of Education Statistics, 2018 (NCES 2020-009), Chapter 2., available at https://nces.ed.gov/fastfacts/display.asp?id=84

```python
from matplotlib import rcParams
#rcParams['font.family'] = 'serif'
#rcParams['font.serif'] = ['Palatino']
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
total = 6606
four = 2832
two = 1528
other = total - (four + two)
```

```python
labels = ['4-year', '2-year', 'Non-degree']# granting']
group_size = [four, two, other]
```

```python
total12 = 7253
four12 = 3026
two12 = 1700
other12 = total12 - (four12 + two12)
```

```python
group_size12 = [four12, two12, other12]
```

```python
# Create a df from the NCES FastFacts data
nces = {'2012':[four12, two12, other12], '2017':[four, two, other], 'names':['4-year', '2-year','Non-degree']}
nces =  pd.DataFrame(nces)
nces.set_index('names', inplace=True)
nces.head()
```

```python
fig, ax = plt.subplots(figsize = (5, 5))
plt.pie(nces['2017'], labels=nces.index, startangle=90)
plt.title('AY 2016-2017', fontsize=16)

# add a circle at the center
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
ax.annotate(total, xy=(.29, .5), fontsize=50, xycoords='axes fraction')
ax.annotate('Post-secondary', xy=(.3, .4), fontsize=16, xycoords='axes fraction')
ax.annotate('institutions', xy=(.35, .33), fontsize=16, xycoords='axes fraction')
plt.tight_layout()
#plt.show()
#plt.savefig('figures/NCESFastFacts2017.png')
```

```python
fig, ax = plt.subplots(figsize = (5, 5))
plt.pie(nces['2017'],labels=nces.index,autopct=lambda p:'{:.0f}'.format(p * total / 100), pctdistance=0.85) #autopct='%1.2f'
plt.title('AY 2016-2017', fontsize=16)

# add a circle at the center
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
ax.annotate(total, xy=(.28, .5), fontsize=50, xycoords='axes fraction')
ax.annotate('Post-secondary', xy=(.29, .4), fontsize=16, xycoords='axes fraction')
ax.annotate('institutions', xy=(.34, .33), fontsize=16, xycoords='axes fraction')
#plt.show()
#plt.savefig('figures/NCESFastFactsAnnotate.png')
```
